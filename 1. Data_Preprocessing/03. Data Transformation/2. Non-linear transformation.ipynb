# Non-linear transformation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#non-normal data
np.random.seed(0)
data = np.random.exponential(size=1000, scale=2)
data = pd.DataFrame(data, columns=['values'])
data.head()
```

```python
data.shape
```

```python
sns.histplot(data['values'], kde=True )
```

## Quantile Transformer

```python
from sklearn.preprocessing import QuantileTransformer

quantile_normal = QuantileTransformer(output_distribution='normal')
```

```python
data['Quantile'] = quantile_normal.fit_transform(data[['values']])
```

```python
sns.histplot(data['Quantile'], kde=True )
```

```python
#import Power Transformer 
from sklearn.preprocessing import PowerTransformer
```

## Box Cox Transformer

```python
pt_boxcox = PowerTransformer(method='box-cox')
```

```python
data['Box-Cox']= pt_boxcox.fit_transform(data[['values']] + 1)
```

```python
sns.histplot(data['Box-Cox'], kde=True )
```

## Yeo-Johnson

```python
pt_yeo_johnson = PowerTransformer(method='yeo-johnson')
```

```python
data['Yeo-Johnson'] = pt_yeo_johnson.fit_transform(data[['values']])
```

```python
sns.histplot(data['Yeo-Johnson'], kde=True )
```
